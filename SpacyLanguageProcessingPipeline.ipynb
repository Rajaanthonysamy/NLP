```python
nlp=spacy.blank("en")
doc=nlp("Captain america ate 100$ of samosa. Then he said I can do this all day.")
```

```python
nlp=spacy.load("en_core_web_sm")
doc=nlp("Captain america ate 100$ of samosa. Then he said I can do this all day.")
```

```python
nlp.pipe_names
```

```python
nlp.pipeline
```

```python
for token in doc:
    print(token, " | ",token.pos_, " | ",token.lemma_)
```

```python
doc =nlp("Tesla Inc is going to acquire twitter for $45 billion")
```

```python
for ent in doc.ents:
    print(ent)
```

```python
for ent in doc.ents:
    print(ent.text,  " | ", ent.label_)
```

```python
for ent in doc.ents:
    print(ent.text,  " | ", ent.label_," | ",spacy.explain(ent.label_))
```

The displacy library is a visualization tool integrated into spaCy, a popular open-source natural language processing (NLP) library for Python.

```python
from spacy import displacy
```

```python
displacy.render(doc,style="ent")
```

```python
doc =nlp("DreamData is started today by us. Our main goal is to become data oriented company")
displacy.render(doc,style="ent")
```

Adding custom components with blank spacy lib

```python
nlp=spacy.blank("en")
doc=nlp("Captain america ate 100$ of samosa. Then he said I can do this all day.")
for ent in doc.ents:
    print(ent.text,  " | ", ent.label_," | ",spacy.explain(ent.label_))
```

```python
source_nlp=spacy.load("en_core_web_sm")
nlp=spacy.blank("en")
nlp.add_pipe("ner",source=source_nlp)
nlp.pipe_names
```

```python
doc=nlp("Tesla Inc is going to acquire Twitter for $45 billion")
for ent in doc.ents:
    print(ent.text,  " | ", ent.label_," | ",spacy.explain(ent.label_))
```

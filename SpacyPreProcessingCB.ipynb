```python
from spacy.symbols import ORTH
nlp.tokenizer.add_special_case("gimme",[
    {ORTH : "gim"},
    {ORTH : "me"}
])
```

```python
doc=nlp("Dr.Strange love pav bhaji of mumbai. Hulk loves chaat of delhi")

for sent in doc.sents:
    print(sent)
```

```python
nlp.add_pipe('sentencizer')
```
